```python
import math
def get_average(records):
    return sum(records) / len(records)
def get_variance(records):
    average = get_average(records)
    return sum([(x - average) ** 2 for x in records]) / len(records)
def get_standard_deviation(records):
    variance = get_variance(records)
    return math.sqrt(variance)
def get_z_score(records):
    avg = get_average(records)
    stan = get_standard_deviation(records)
    scores = [(i-avg)*100/stan for i in records]
    return scores
```

Forestry

```python
df2.iloc[3,1] = 180
data_Capacity = df2[['Region_CN','Forestry.1']]
attr_Capacity = data_Capacity[['Region_CN']]
attr = attr_Capacity.values.tolist()
attr = reduce(operator.add, attr)
value_Capacity = data_Capacity[['Forestry.1']]
value = value_Capacity.values.tolist()
value = reduce(operator.add, value)
# Normalize data to 1-100
Value = get_z_score(value)
# Value = [(x - min(value))*100/(max(value) - min(value)) for x in value]
map = Map("Forestry", width=1200, height=600)
map.add("", attr, Value, maptype='china', is_visualmap=True, visual_text_color='#000')
map.show_config()
map.render()
```
